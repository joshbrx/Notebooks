```python
import numpy as np
import matplotlib.pyplot as plt
import random
```

```python
# Constantes
hbar = 1.054571817e-34  # J*s
k_B  = 1.380649e-23     # J/K

# Valores proporcionados
T = 273       # K
m = 4.65e-26  # kg
```

\begin{equation}\label{eq:}
n(u)=\left(\frac{N\lambda_D^3m^3}{2\pi^2\hbar^3}\right)u^2e^{-\frac{u^2m}{2k_B T}}
\end{equation}

con $\lambda_D=\lambda_D(T)=\left(\frac{2\pi\hbar^2}{m k_B T}\right)^{1/2}$

```python
l_D = lambda T: ((2*np.pi*hbar**2)/m*k_B*T)**(1/2)
l_D(T)
```

```python
def n(u, N):
    t1 = (N*(l_D(T)**3)*(m**3)/(2*np.pi**2*hbar**3))
    t2 = u**2 
    t3 = np.exp(-u**2*m/(2*k_B*T))
    return t1*t2*t3
```

```python
U = np.linspace(0,1000); N = 10e23

plt.figure(figsize=(12,5))
plt.plot(U, n(U,N))
plt.xlabel(r"u"); plt.ylabel(r"n(u)")
plt.title(r"Distribución de Maxwell-Boltzmann con $T=273\ K$ y $m=4.65\dot 10^{-26}\ kg$")
plt.show()
```

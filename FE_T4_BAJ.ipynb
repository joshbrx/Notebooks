- _Movimiento browniano en 2D_

Usa las soluciones $X(t) = N^t_{0,x}(0, 1)\sqrt{\delta^2 t}$ y $Y(t) = N^t_{0,y}(0, 1)\sqrt{\delta^2 t}$ y el método en la sección 6.4 (del libro) para generar y graficar el camino de una partícula browniana en el plano $xy$. Supón que las normales unitarias $N^t_{0,x}(0, 1)$ y $N^t_{0,y}(0, 1)$ son estadísticamente independientes.


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
import random
```

```python
def mov_browniano(x0, N, dt, delta):
    x0 = np.array(x0)
    r = norm.rvs(size=x0.shape + (N,), scale=delta*np.sqrt(dt))
    
    temp = np.cumsum(r, axis=-1)
    temp += np.expand_dims(x0, axis=-1)
    return temp
```

```python
# Condiciones iniciales
x0 = [0,0];
N = 1000; 
dt = 0.2;
delta = 0.5

MB = mov_browniano(x0, N, dt, delta)
```

```python
plt.figure(figsize=(8,7))
plt.scatter(MB[0], MB[1], s=5, c="orange")
plt.scatter(MB[0,0], MB[1,0], s=100, c="Blue", label="Inicio")
plt.scatter(MB[0,-1], MB[1,-1], s=100, c="red", label="Final")
plt.plot(MB[0], MB[1], linewidth=0.5)
plt.xlabel(r"$X(t)$"); plt.ylabel(r"$Y(t)$"); plt.title(r"Movimiento Browniano con $N=1000$ pasos")
plt.legend(); plt.grid(); plt.show()
```

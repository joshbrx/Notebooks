1. Desmuestre que las siguiente ecuación tiene una solución en los inetervalos dados:

$$x\cos x -2 x^2+3 x-1 = 0,\qquad\,\,\,[0.2,0.3] \quad\, y \quad[1.2,1.3]$$

```python
# Primero definimos nuestra variable "x" en el el dominio

x1 = np.linspace(0.2, 0.3)
x2 = np.linspace(1.2, 1.3)
x =  np.linspace(-10, 10)
zeros1 = np.zeros(x1.shape)
zeros2 = np.zeros(x2.shape)

# Ahora definimos la función que se nos dió

f = lambda x: x*np.cos(x)-2*x**2+3*x-1

y1 = f(x1); y2 = f(x2); y = f(x)

#Finalmente graficamos la variable "y" sobre el dominio de sus respectivas "x"

plt.figure(figsize = (15,5))
plt.subplot(1,2,1)          # Primera gráfica (y1 vs x1)
plt.plot(x1, y1)
plt.plot(x1, zeros1)
plt.xlabel("x"); plt.ylabel("y") 
plt.title(r"x $\in [0.2, 0.3]$ ")

plt.subplot(1,2,2)          # Segunda gráfica (y2 vs x2)
plt.plot(x2, y2)
plt.plot(x2, zeros2)
plt.xlabel("x"); plt.ylabel("y") 
plt.title(r"x $\in [1.2, 1.3]$ ")

plt.show()
```

7. Sea $f(x)=x^3$  

    a. Determinite el segundo polinomio de Taylor $P_2(x)$ en torno a $x_0=0$  
    b. Calcule $R_2(0.5)$ y el error real al usar $P_2(0.5)$ para aproximar $f(0.5)$  
    c. Repita el inciso (a) usando $x_0=1$  
    d. Repita el inciso (b) con el polinomio del inciso (c)  

Primero escribimos la definición de las series de Taylor, que es dada $f\in C^n[a,b]$, que $f^{(n+1)}$ existe en $[a,b]$ y $x_0\in[a,b]$. Para cada $x\in [a,b]$, existe un número $\xi(x)$ entre $x_0$ y $x$ tal que

$$f(x)=P_n(x)+R_n(x)$$ donde

$$P_n(x)=\sum_{k=0}^{n}\frac{f^{(k)}(x_0)}{k!}(x-x_0)^k\qquad\qquad  y \qquad\qquad  R_n=\frac{f^{(n+1)}\xi(x)}{(n+1)!}(x-x_0)^{(n+1)}\$$

Dicho lo anterior, **podemos comenzar!**

```python
# Primero definimos la función f(x)=x**3
f = lambda x: x**3

# Ahora ya que f'(0)=f''(x)=0, y finalmente f'''(x)=6,
# definimos la serie de Taylor alrededor del 0
f_2p = lambda x: 0

# Calculamos R2
R2 = f(0.5)-f_2p(0.5)
print(fr"R2 = {R2}, entonces podemos aproximar a f(0.5) = {f_2p(0.5) + R2}")
```

```python
# Reperimos para x_0=1, entonces f(1)=1

# Ahora ya que f'(1)=3, f''(1)=6, y finalmente f'''(x)=6,
f_2p = lambda x: 1 + 3*(x-1) + 3*(x-1)**2 + (x-1)**3 

# Calculamos R2
R2 = f(0.5)-f_2p(0.5)
print(fr"R2 = {R2}, entonces podemos aproximar a f(0.5) = {f_2p(0.5) + R2})")
```

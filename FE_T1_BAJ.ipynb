### Ejercicio 2: _Modos de falla independientes_

Un sistema consiste de $n$ componentes separados, donde cada uno falla independientemente de los otros con probabilidad $P_i$, donde $i = 1...n$. Dado que cada componente falla o no falla, la probabilidad de que el $i$-esimo componente no falle es $1 − P_i$.

__a)__ Supón que los componentes están conectados en paralelo de tal manera que la falla de todos los componentes es necesaria para que el sistema falle. ¿Cuál es la probabilidad de que el sistema falle?, y ¿cuál es la probabilidad de que el sistema funcione (no falle)?

Por hipótesis tenemos que las componentes del sistema están conectadas ___en paralelo___ por lo que el sistema entero fallará solamente solo cuando fallen todas las componentes al mismo tiempo. Es decir que la probabilidad del sistema $P_{s,par}$ en el caso $x_0:\text{sí falla}$ y el caso $x_1:\text{no falla}$ , es la siguiente

\begin{equation*}
    P_{s,par}(x_0)=\prod_{i=1}^nP_i,\qquad\ P_{s,par}(x_1)=1-\prod_{i=1}^nP_i
\end{equation*}


-------

__b)__ Supón que los componentes están conectados en serie, de tal manera que la falla de un componente causa la falla de todo el sistema. ¿Cuál es la probabilidad de que el sistema falle? (Pista: Primero encuentra la probabilidad de que todos los componentes funcionen.)

En esta nueva configuración tenemos que las componentes del sistema están conectadas ___en serie___ por lo que el sistema entero fallará cuando al menos una de las componentes. Es decir que la probabilidad del sistema $P_{s,serie}$ en el caso $x_0:\text{sí falla}$ y el caso $x_1:\text{no falla}$ , es la siguiente

\begin{equation*}
    P_{s,serie}(x_0)=1-\prod_{i=1}^n(1-P_i),\qquad\ P_{s,serie}(x_1)=\prod_{i=1}^n(1-P_i)
\end{equation*}

### Ejercicio 3: _Caminantes aleatorios... o borrachos_

Supón que tienes un caminante aleatorio ”con duda”, es decir, que en una iteración dada además de tener la posibilidad de moverse una unidad tanto hacia adelante como hacia atrás, también puede decidir no moverse. Calcula la varianza de la posición del caminante como función del número de pasos, $n$, tomados en los siguientes casos:

a) $P(x_i = −1) = \frac{1}{3},\quad\ P(x_i = 0) = \frac{1}{3},\quad\ P(x_i = 1) = \frac{1}{3} $

b) $P(x_i = −1) = \frac{1}{4},\quad\ P(x_i = 0) = \frac{1}{2},\quad\ P(x_i = 1) = \frac{1}{4} $

c) $P(x_i = −1) = \frac{1}{4},\quad\ P(x_i = 0) = \frac{1}{4},\quad\ P(x_i = 1) = \frac{1}{2}$

Recuerda que cada una de estas caminatas está dada por la siguiente suma de variables aleatorias

\begin{equation}
Y_n = X_1 + X_2 + \dots + X_n = \sum_{j=1}^n X_j 
\end{equation}

donde a su vez, cada $X_i$ es una variable aleatoria que puede tomar los valores $x_1 = −1$, $x_2 = 0$ y $x_3 = 1$ con las probabilidades dadas por cada uno de los incisos.

------
Dado que la varianza esta definida como $\sigma_X^2=\text{VAR}\{X\}=\langle X^2\rangle-\langle X\rangle^2$

__a)__ Primero calculamos $\langle X_i\rangle$ y $\langle X_i^2\rangle$ por separado

\begin{align*}
\langle X_i\rangle = \sum_{j=1}^n x_{i,j}P_j= &(-1+0+1)\cdot 1/3=0 \\
\langle X_i^2\rangle = \sum_{j=1}^n x_{i,j}^2P_j= &\left[(-1)^2+0^2+1^2\right]\cdot 1/3=2/3
\end{align*}

Entonces 

\begin{equation*}
\sigma_{X_i}^2=\text{VAR}\{X_i\} =\langle X_i^2\rangle-\langle X_i\rangle^2=2/3-0=2/3
\end{equation*}

Y finalmente para $Y=\sum_{i=1}^nX_i$

\begin{align*}
\sigma_Y^2 = \text{VAR}\{Y\}=\text{VAR}\{\sum_{i=1}^nX_i\}=\sum_{i=1}^n \text{VAR}\{X_i\}=n\cdot\text{VAR}\{X_i\}=\frac{2n}{3} \qquad\ \boxed{\therefore \sigma_Y^2 =\frac{2n}{3}} \quad\text{a)}
\end{align*} 

-----------

__b)__ De manera análoga, se calcula $\langle X_i\rangle$ y $\langle X_i^2\rangle$ por separado

\begin{align*}
\langle X_i\rangle = \sum_{j=1}^n x_{i,j}P_j &=\left(-1\cdot1/4+0\cdot1/2+1\cdot1/4\right)=0 \\
\langle X_i^2\rangle = \sum_{j=1}^n x_{i,j}^2P_j &=\left[(-1)^2\cdot1/4+0\cdot1/2+1^2\cdot1/4\right]=1/2
\end{align*}

Entonces 

\begin{equation*}
\sigma_{X_i}^2=\text{VAR}\{X_i\} =\langle X_i^2\rangle-\langle X_i\rangle^2=1/2-0=1/2
\end{equation*}

Y finalmente para $Y=\sum_{i=1}^nX_i$

\begin{align*}
\sigma_Y^2 = \text{VAR}\{Y\}=\text{VAR}\{\sum_{i=1}^nX_i\}=\sum_{i=1}^n \text{VAR}\{X_i\}=n\cdot\text{VAR}\{X_i\}=\frac{n}{2} \qquad\ \boxed{\therefore \sigma_Y^2 =\frac{n}{2}} \quad\text{b)}
\end{align*} 

-----------

__c)__ Por último, se calcula $\langle X_i\rangle$ y $\langle X_i^2\rangle$ por separado

\begin{align*}
\langle X_i\rangle = \sum_{j=1}^n x_{i,j}P_j &=\left(-1\cdot1/4+0\cdot1/4+1\cdot1/2\right)=1/4 \\
\langle X_i^2\rangle = \sum_{j=1}^n x_{i,j}^2P_j &=\left[(-1)^2\cdot1/4+0\cdot1/4+1^2\cdot1/2\right]=3/4
\end{align*}

Entonces 

\begin{equation*}
\sigma_{X_i}^2=\text{VAR}\{X_i\} =\langle X_i^2\rangle-\langle X_i\rangle^2=3/4-(1/4)^2=11/16
\end{equation*}

Y finalmente para $Y=\sum_{i=1}^nX_i$

\begin{align*}
\sigma_Y^2 = \text{VAR}\{Y\}=\text{VAR}\{\sum_{i=1}^nX_i\}=\sum_{i=1}^n \text{VAR}\{X_i\}=n\cdot\text{VAR}\{X_i\}=\frac{11n}{16} \qquad\ \boxed{\therefore \sigma_Y^2 =\frac{11n}{16}} \quad\text{c)}
\end{align*}

__a)__ Gráfica 20 de las (1000) listas junto con la desviación estándar de la expresión analítica que

```python
volados_a = [];
pos_a = np.zeros((1000,1000));
for i in range(1000):
    volados_a.append(np.array(random.choices([-1, 0, 1], weights = (1/3, 1/3, 1/3), k=1000)))
    for k in range(1000):
        pos_a[i][k]=(np.sum(volados_a[i][0:k+1]))
```

```python
f = lambda x: np.sqrt(2*x/3);
X = np.linspace(0,1000,1000);

plt.figure(figsize=(14,5)) 
for i in range(20):
    plt.plot(np.arange(1000),pos_a[i])

plt.plot(X,f(X), "r", label=r"$\sigma=\sqrt{\frac{2n}{3}}$"); plt.plot(X,-f(X), "r")

plt.xlabel(r"Número de paso"); plt.ylabel(r"Posición"); plt.title("Desplazamiento de los caminantes aleatorios")
plt.legend(); plt.grid(); plt.show()
plt.show();
```

```python
# Importamos las librerias para hacer webscraping
import requests
from bs4 import BeautifulSoup
import re
```

### Web Scraping con Python

Para hacer webscraping necesitamos primero localizar la pagina de nuestro interés para luego poder tener el codigo en html y poder hacer un proceso de data mining más eficiente

```python
# Hacemos una lista de las páginas que necesitamos y definimos la función "añadir_url"
urls = ['http://www.fciencias.unam.mx/docencia/horarios/20211/1081/609',
 'http://www.fciencias.unam.mx/docencia/horarios/20211/1081/715',
 'http://www.fciencias.unam.mx/docencia/horarios/20211/1081/716',
 "http://www.fciencias.unam.mx/docencia/horarios/20211/1081/717"]
```

```python
# Salvamos la lista "urls" como un archivo .txt

with open("urls.txt", "w") as output:
    output.write(str(urls))
```

```python
# Volvemos a abrir el archivo para trabajar con los links

my_file = open("urls.txt", "r")
content = my_file.read().replace("[","").replace("]","").replace("'","").replace(" ","")
url_list = content.split(","); my_file.close()
url_list
```

```python
regex1 = re.compile("(\D+),")
regex2 = re.compile("\nProfesor\n(.+)\n(.+)\n\n(\d+ a \d+)\n")
regex3 = re.compile("\nLaboratorio\n(.+)\n(.+)\n\n(\d+ a \d+)\n")
```

```python
soups = []; materias = []; datos = []
for i in range(len(url_list)):
#     print(url_list[i])
    soups.append(BeautifulSoup(requests.get(url_list[i]).content, "html.parser"))
    for j in soups[i].select("div#contenedor div#contenido div.info-estudiantes div#info-contenido h2"):
        materias.append(regex1.findall(j.get_text())[0])
        print("\n" + "ow"*50 + "\n" + j.get_text())
    for k in soups[i].select("div#contenedor div#contenido div.info-estudiantes div#info-contenido table"):
        print(materias[-1], k.get_text(), "\n------------------------\n")
#         print("XD", regex2.findall(k.get_text()) , "XD" )
        s = regex2.findall(k.get_text())
        if s == []: s = regex3.findall(k.get_text())
        if len(s) != 3: s = s[0]
        datos.append([materias[-1]] + list(s))
        
```

```python
datos[0]
```

```python
#Ahora convertimos la variable datos en una tabla con pandas
import pandas as pd

data = pd.DataFrame(datos, columns=["Materia", "Profesor", "D", "H"])
data["Calif"]=0.0
data.head()
```

```python
#Definimos una función para añadir una calificación según "MisProfesores.com" a cada profesor
def calificar(data):
    for k in range(len(data["Profesor"])):
        data.at[k, 'Calif'] = float(input(data["Profesor"][k] +": "))
calificar(data)
```

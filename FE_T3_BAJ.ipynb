### Ejercicio 2.

Genera $n = 10$ realizaciones distintas de la variable $X_1$ y agrégalas a una lista. Reporta
explícitamente 10 listas (10 caminantes) distintas con $n = 10$ realizaciones de la variables $X_1$ (10
pasos).

```python
def X(pasos, listas):
    tabla = np.array([random.choice([-1, 1]) for n in range(pasos*listas)])
    tabla = tabla.reshape(pasos, listas)
    return tabla.astype("float32")
```

```python
datos = X(15, 10)

datos_c = []
for i in range(datos.shape[1]):
    datos_c.append("Lista " + str(int(i+1)))

pd.DataFrame(data=datos, columns=datos_c)
```

### Ejercicio 3.

Para cada una de estas listas obtén el promedio y repórtalos.

Lo que hiciste en el punto 3 fue obtener 10 realizaciones de la variable aleatoria $\overline{X}=\frac{X_1+X_2+...+X_n}{n}$,
donde cada una de las variables $X_i$ que la constituyen es idéntica a aquella que construiste en el
inciso 1. Nota que como discutimos en clase, $\overline{X}$ no es el promedio de una de las $X_i$’s, sino una
**variable aleatoria** cuyo promedio es igual al promedio de $\langle X_i\rangle$. Por eso, el promedio de cada lista
no es igual a cero (o por lo menos no de todas). Para asegurar que el promedio fuera estrictamente
igual a 0, $n$ tendría que ser infinito.

```python
def prom(datos):
    return np.average(datos, axis=0)
```

```python
promedios = prom(datos)
promedios = promedios.reshape(10,1)

promedios_c = []
for i in range(promedios.shape[0]):
    promedios_c.append("Prom " + str(int(i+1)))

pd.DataFrame(data=promedios.T, columns=promedios_c)
```

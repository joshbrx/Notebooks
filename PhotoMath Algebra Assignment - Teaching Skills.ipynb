__1.__ You order __two burritos__ and __a side dish__. The side dish costs \\$2.50.
You pay 8\% sales tax and leave a \$3 tip. You pay a total of \\$13.80. 

&nbsp;&nbsp;&nbsp; How much does one burrito cost? Explain each step.

-------------

As the problem states, you ordered $2$ burritos and $1$ side dish (let's assing them the letter $x$ and $y$ respectively) and we paid $8\%$ of taxes and a $\$3$ tip. In the end, you paid a total of $\$13.80$.

With all that info, we have the following equation system (with $y=2.50$)

$$\left[1+\frac{8}{100}\right](2x+y)+3=13.8$$

first we make the right side of the equation equal $0$ by subtracting $13.8$ in each side

$$\Rightarrow \left[1+\frac{8}{100}\right](2x+y)+3 - 13.8=13.8 - 13.8 =0$$

simplifying numerical values

$$\Rightarrow(1.08)(2x+y) -10.8 =(2x+y)(1.08) -10(1.08) =0$$

factoring out and dividing by $(1.08)$

$$\Rightarrow [2x+y-10](1.08) =0,\qquad \Rightarrow 2x+y-10 =0$$

now we can solve easily for $x$

$$\Rightarrow 2x=10-y,\qquad \Rightarrow x=\frac{10-y}{2}=5-\frac{y}{2}$$

substituting $y=2.5$

$$x=5-(2.5)/2=5-1.25=3.75,\qquad \therefore x=3.75$$

#### $$\text{Therefore, the value of each burrito is $\$3.75$}$$

__2.__ The differences between the terms of the geometric sequence are known:

$$a_2-a_1=-6$$
$$a_3-a_2=12$$

&nbsp;&nbsp;&nbsp;
Find the sum of the first ten terms of that sequence.

-------------------

First, let's remember some things about geometric sequences:

Geometric sequences have a common ratio $r$ and the formula for the $n^{th}$ term is

$$a_n=a_1(r)^{n-1}$$

where $a_n$ is the $n^{th}$ term of the sequence, $a_1$ is the first term of the sequence and $r$ is the common ratio.



Using the previous formula, we can solve for $r$ in terms of $a_1$ and $a_n$

$$a_n=a_1(r)^{n-1},\qquad \Rightarrow r^{n-1}=\frac{a_n}{a_1},\qquad \therefore r=\left(\frac{a_n}{a_1}\right)^{\frac{1}{n-1}}$$

and for $n=2$ and $n=3$

$$r=\frac{a_2}{a_1},\qquad r=\sqrt{\frac{a_3}{a_1}}$$

Now we need to rearrange the equations beggining to find the ratio. 

Solving the first equation

$$a_2-a_1=-6,\qquad \Rightarrow a_2=a_1-6,\qquad$$

dividing by $a_1$
$$\Rightarrow \frac{a_2}{a_1}=\frac{a_1-6}{a_1}=1-\frac{6}{a_1},\qquad \therefore \boxed{r=1-\frac{6}{a_1}}$$

Analogously for the second equation

$$a_3-a_2=12,\qquad \Rightarrow a_3=12 + a_2=12+(a_1-6)$$

dividing by $a_1$
$$\Rightarrow \frac{a_3}{a_1}=\frac{a_1+6}{a_1}=1+\frac{6}{a_1},\qquad \therefore \boxed{r=\sqrt{\frac{a_3}{a_1}}=\sqrt{1+\frac{6}{a_1}}}$$



Since $r$ has two values, both of them have to be equal

$$\sqrt{1+\frac{6}{a_1}}=1-\frac{6}{a_1}$$

squaring both sides
$$1+\frac{6}{a_1}=\left(1-\frac{6}{a_1}\right)^2=1-2\left(\frac{6}{a_1}\right)+\frac{6^2}{a_1^2}$$


simplifying and grouping terms
$$\Rightarrow 3\left(\frac{6}{a_1}\right)=\frac{6}{a_1}\left(\frac{6}{a_1}\right)$$

dividing by $(6/a_1)$
$$\Rightarrow 3=\frac{6}{a_1},\qquad \Rightarrow 3a_1=6,\qquad \therefore \boxed{a_1=2}$$

Finally, we can solve $r$ with the previous equation for $n=2$

$$\Rightarrow r=1-\frac{6}{a_1}=1-\frac{6}{(2)}=1-3=-2,\qquad \therefore \boxed{r=-2}$$

So, the formula for this particular case is 

$$a_n=2(-2)^{n-1}$$

&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp; &nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;  &nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;

__3.__ Obtain the graph of the parabola $y=-x^2+4x$ using transformations of the graph of $\hat{y}=(\hat{x})^2$. 

&nbsp;&nbsp;&nbsp; 
Explain each step.

-------------------

First, we need to convert the parabola $y=-x^2+4x$ into the standard form. 

$$4p(x-h)=\pm(y-k)^2$$

so let's add a $0$ in the equation
$$y=(-x^2+4x)+0=(-x^2+4x)-4+4=-(x^2-4x+4)+4$$

completing the square
$$\Rightarrow y=-(x-2)^2+4,\qquad \therefore \boxed{y-4=-(x-2)^2}$$

By inspection we can see the center $(h,k)$ of the parabola is $(2,4)$ and $p=1/4$, and following the definition of coordinates transformation

$$x'=x-h,\qquad y'=y-k$$

substituting the values of $h$ and $k$
$$x'=x-2,\qquad y'=y-4$$

replacing in the parabola formula
$$\Rightarrow y'=-(x')^2$$

Finally, we need to change the sign of the variable $y'$ to match the parabola $\hat{y}=\hat{x}^2$

$$\hat{x}=x'=x-2,\qquad \hat{y}=-y'=-(y-4)$$

$$\therefore \boxed{\hat{x}=x-2,\quad \hat{y}=4-y}$$

```python
# You don't need to understand this code, it's for visualization purpuses
import numpy as np
import matplotlib.pyplot as plt

x_hat = np.linspace(-3,3)
y_hat = (lambda x: x**2)(x_hat)

x = (lambda x: x + 2)(x_hat)
y = (lambda y: 4 - y)(y_hat)

plt.plot(x, y, label=r"$-(y-4)=(x-2)^2$")
plt.plot(x_hat, y_hat, label=r"$y = x^2$")
plt.scatter([0,2],[0,4],)
plt.legend()
plt.show()
```

As you can see in the graph above, the center $(0,0)$ matches with $y=x^2$, and the center $(2,4)$ matches with $-(y-4)=(x-2)^2$

__5.__ Which of the six trigonometric functions are odd? And which are even? 

&nbsp;&nbsp;&nbsp;
Justify the answers using trigonometric identities and graphs.

-------------------

We need to show $f(x)=f(-x)$ for an even function, and $g(x)=-g(-x)$ for an odd function.

- For $\sin x$ (and $\csc x = 1/\sin x$):

We can see $\sin x$ is odd in the graph below, and since $\sin x$ is odd, it holds the following

$$\sin x = -\sin (-x)\qquad \Rightarrow \csc x=\frac{1}{\sin x} = -\frac{1}{\sin (-x)}=-\csc(-x)$$

Therefore $\csc x$ is an odd function.

```python
x = np.linspace(-np.pi+0.02,-0.02+np.pi, 300)
fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2, figsize=(14,3.5))
ax1.plot(x,(lambda x: np.sin(x))(x)); ax1.plot(x,x*0); ax1.axvline(x=0); ax1.set_title("Sine of x")
ax2.plot(x,(lambda x: 1/np.sin(x))(x)); ax2.plot(x,x*0); ax2.axvline(x=0); ax2.set_title("Cosecant of x")
plt.show()
```

- For $\cos x$ (and $\sec x = 1/\cos x$):

We can see $\cos x$ is even in the graph below, and since $\cos x$ is even, it holds the following

$$\cos x = \cos (-x)\qquad \Rightarrow \sec x=\frac{1}{\cos x} = -\frac{1}{\cos (-x)}=-\sec(-x)$$

Therefore $\sec x$ is an even function.

```python
x = np.linspace(-np.pi+0.02,-0.02+np.pi, 300)
fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2, figsize=(14,3.5))
ax1.plot(x,(lambda x: np.cos(x))(x)); ax1.plot(x,x*0); ax1.axvline(x=0); ax1.set_title("Cosine of x")
ax2.plot(x,(lambda x: 1/np.cos(x))(x)); ax2.plot(x,x*0); ax2.axvline(x=0); ax2.set_title("Secant of x")
plt.show()
```

- For $\tan x$ (and $\cot x = 1/\tan x$):

Since $\tan x=\frac{\sin x}{\cos x}=\sin x(\sec x)$ and the functions $\sin x$ and $\sec x$ are odd and even functions respectively, we know an odd function times an even function equals an odd function.

Therefore $\tan x$ is an odd function. (Analogously for $\cot x) 

```python
x = np.linspace(-np.pi+0.02,-0.02+np.pi, 264)
fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2, figsize=(14,3.5))
ax1.plot(x,(lambda x: np.tan(x))(x)); ax1.plot(x,x*0); ax1.axvline(x=0); ax1.set_title("Tangent of x")
ax2.plot(x,(lambda x: 1/np.tan(x))(x)); ax2.plot(x,x*0); ax2.axvline(x=0); ax2.set_title("Cotangent of x") 
plt.show()
```
